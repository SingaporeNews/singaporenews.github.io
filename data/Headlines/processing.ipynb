```python
import pandas as pd
import nltk as nltk
from nltk.corpus import stopwords
import glob
import scipy as scipy
import collections
```

```python
allFiles = glob.glob("*.txt")
df = pd.DataFrame()
#list = []
for files in allFiles:
    frame = pd.read_table(files)
    df = df.append(frame, ignore_index=True)

df = df.dropna()
```

```python
df['tokens'] = df.Headline.apply(lambda x: nltk.word_tokenize(x.lower()))
```

```python
df['tokens'] = df.tokens.apply(lambda x: [word.translate(None, "!@#$;':-*^&") for word in x])
```

```python
replacements = [
                ("s'pore","singapore"),
                ("spore", "singapore"),
                ("govt", "government"),
                ("pts", "points"),
                ]
```

```python
def version_replace(x, replacements):
    for item in replacements:
        if x == item[0]:
            x = item[1]
        else:
            x = x
    return x
```

```python
english_stops = set(stopwords.words('english'))
english_stops.discard('up')
english_stops.discard('down')
english_stops.add('...')
english_stops.add("'s")
english_stops.add('mr.')
english_stops.add("n't")
english_stops.add("'")
english_stops.add(":")
```

```python
df['tokens'] = df['tokens'].apply(lambda x: [word for word in x if word not in english_stops and len(word) > 1])
df['tokens'] = df['tokens'].apply(lambda x: [version_replace(word, replacements) for word in x])
```

#### Co-occurrence Matrix

```python
def cooccurence_matrix_corpus(corpus):
    matrix = collections.defaultdict(lambda : collections.defaultdict(int))
    
    for corpora in corpus:
        for i in xrange(len(corpora)-1):
            for j in xrange(i+1, len(corpora)):
                word1, word2 = [corpora[i], corpora[j]]
                matrix[word1][word2] += 1
                matrix[word1][word2] += 1
    return matrix
```

```python
ccm = cooccurence_matrix_corpus(df['tokens'])
```

```python
ccm_dict = []
for node in ccm:
    datadict = {}
    if type(node) == str:
        datadict['name'] = node
        word_list = []
        #datadict[node] = []
        counter = 0;
        for item in ccm[node].items():
            #d = {}
            #d['name'] = item[0]
            #d['size'] = item[1]
            #datadict[node].append(d)
            counter += item[1]
            word_list.append(item[0])
        datadict['links'] = word_list
        datadict['size'] = counter
    ccm_dict.append(datadict)
```
