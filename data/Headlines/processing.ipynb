```python
import pandas as pd
import nltk as nltk
from nltk.corpus import stopwords
import glob
```

```python
allFiles = glob.glob("*.txt")
df = pd.DataFrame()
#list = []
for files in allFiles:
    frame = pd.read_table(files)
    df = df.append(frame, ignore_index=True)

df = df.dropna()
```

```python
df['tokens'] = df.Headline.apply(lambda x: nltk.word_tokenize(x.lower()))
```

```python
df['tokens'] = df.tokens.apply(lambda x: [word.translate(None, "!@#$;':-*^&") for word in x])
```

```python
replacements = [
                ("s'pore","singapore"),
                ("govt", "government"),
                ("pts", "points"),
                ]
```

```python
def version_replace(x, replacements):
    for item in replacements:
        if x == item[0]:
            x = item[1]
        else:
            x = x
    return x
```

```python
english_stops = set(stopwords.words('english'))
english_stops.discard('up')
english_stops.discard('down')
english_stops.add('...')
english_stops.add("'s")
english_stops.add('mr.')
english_stops.add("n't")
english_stops.add("'")
english_stops.add(":")
```

```python
df['tokens'] = df['tokens'].apply(lambda x: [word for word in x if word not in english_stops and len(word) > 1])
df['tokens'] = df['tokens'].apply(lambda x: [version_replace(word, replacements) for word in x])
```
